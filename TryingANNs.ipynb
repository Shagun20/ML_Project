```python
df = pd.read_csv("/content/drive/MyDrive/ML project/weatherAUS-noDir.csv")
```

```python
df = df.sample(frac=1)
```

```python
x = list(df.columns).index("RainTomorrow")
ls= list(df.columns[:x])+list(df.columns[x+1:])
ls.append("RainTomorrow")
df = df[ls]
df.head()
del x
```

```python
X=df[df.columns[:-1]]
y = df[df.columns[-1]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

### Easy MLP classifier test accuracy

```python
def try_lr(activ, lr_num):
  clf = MLPClassifier(hidden_layer_sizes=(256, 32), random_state=21, batch_size = 100, activation= activ, learning_rate_init = lr_num)
  losses = []
  val_losses = []

  for i in range(30):
    if i%5 == 0:
      print(f"Epoch: {i}")
    clf.partial_fit(X_train, y_train, np.unique(y))
    losses.append(clf.loss_)
    predictions = clf.predict_proba(X_test)
    val_losses.append(log_loss(y_test, predictions))
  
  plt.plot(losses)
  plt.plot(val_losses)
  plt.title('Plot History: Model Loss')
  plt.ylabel('Loss')
  plt.xlabel('Epoch')
  plt.legend(['Train', 'Test'], loc='upper left')
  plt.show()

  print("Best val loss", clf.best_loss_)
```

```python
def try_lr(activ='logistic', lr = 0.001):
  clf = MLPClassifier(hidden_layer_sizes=(256, 32), random_state=21, batch_size = 100, activation= activ, learning_rate_init = lr)
  clf.fit(X_train, y_train, )
  y_pred = clf.predict(X_test)
  print("Precision:", precision_score(y_test, y_pred))
```
